## Basic Plot

```python
a = np.array(data['Time'])
b = np.array(data['Amount'])
print(plt.plot(a,b))
print(plt.xlabel('Amount'))
print(plt.ylabel('Time'))
print(plt.title('Amount VS Time'))
```

## Bar Plot

```python
x = np.array(data.head()['Time'])
y = np.array(data.head()['Amount'])
print(plt.bar(x,y))
print(plt.xlabel('Amount'))
print(plt.ylabel('Time'))
print(plt.title('Amount VS Time'))
```

## Histogram

```python
p = np.array(data.head(30)['Time'])
print(plt.hist(p))
print(plt.xlabel('Amount'))
print(plt.ylabel('Time'))
print(plt.title('Amount VS Time'))
```
